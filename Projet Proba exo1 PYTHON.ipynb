```python
from matplotlib.pyplot import*
from math import*
from numpy import*
from numpy.random import*
from scipy.stats import*
import scipy.stats
```

```python
N=5000
n=100
Y=10*rand(1,n)
Y
```

```python
yn=sum(Y)/n
yn
```

```python
Y=np.zeros((N,n))
for i in range(0,N):
    Y[i]=10*rand(1,n)
Y
```

```python
yn=np.zeros(N)
for i in range(0,N):
    yn[i]=sum(Y[i])/n
yn
```

```python
figure(figsize=(10,10))
hist(yn,bins=linspace(min(yn),max(yn),50),density=True);
title('Histogramme des 5000 r√©alisation de $\overline{y_n}$',fontsize=30,color="blue")
```

```python
N=10000
n=10
zn=np.zeros(N)
for i in range(0,N-1):
    Y=10*rand(1,n)
    zn[i]=sum(Y)
yn=zn/n
figure(figsize=(10,10))
hist(yn,bins=linspace(min(yn),max(yn),50),density=True);
title('Histogramme des 10000 r√©alisation de $\overline{Y}_n$ avec $n=10$',fontsize=30,color="blue")
```
