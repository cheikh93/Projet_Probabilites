```python
from matplotlib.pyplot import*
from math import*
from numpy import*
from numpy.random import*
from scipy.stats import*
import scipy.stats
```

```python
n=100
Y=np.random.normal(2,2,n)
Y
```

```python
yn=sum(Y)/n
yn
```

```python
zn=sqrt(n)*(yn-2)/2
```

```python
N=5000
Y=np.zeros((N,n))
for i in range(0,N):
    Y[i]=np.random.normal(2,2,n)
Y
```

```python
yn=np.zeros(N)
for i in range(0,N):
    yn[i]=sum(Y[i])/n
yn
```

```python
zn=sqrt(n)*(yn-2)/2
zn
```

```python
figure(figsize=(10,10))
hist(zn,bins=linspace(min(zn),max(zn),50),density=True);
title('Histogramme des 5000 r√©alisation de $\overline{Z}_n$ avec $n=100$',fontsize=30,color="blue")
x=linspace(min(zn),max(zn),50)
plot(x,scipy.stats.norm.pdf(x,0,1),lw=5)
```
